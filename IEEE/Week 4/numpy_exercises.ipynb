<span style = "font-size:20px"> **(8) Flat the following array.**</span><br>

````python
arr = np.array([[i, 2*i, 3*i] for i in range(3, 300)])
flat = arr.flatten()   # or arr.ravel()

print("Before shape:", arr.shape)
print("After shape: ", flat.shape)
print(flat[:9])
````

<span style = "font-size:20px"> **(9) Transpose the following array.**</span><br>

````python
arr = np.array([[[[j, j+1] for j in range(5)]] for i in range(3)])
t = arr.T

print("Before shape:", arr.shape)  
print("After shape: ", t.shape)    
````

##  NumPy Dimensions – Questions

1. What is the shape of `a` and `b`?  

2. What does `(3,)` mean in NumPy? Why is there a comma?  

3. How many dimensions does each array have?  

4. What does each dimension (axis) represent in `b`?  

5. Is `(3,)` the same as `(3, 1)`? Explain the difference.  

````python
import numpy as np

a = np.array([10, 20, 30])
b = np.array([[1,2],[3,4],[5,6]])

print("a.shape:", a.shape)    # (3,)
print("b.shape:", b.shape)    # (3, 2)
print("a.ndim: ", a.ndim)     # 1
print("b.ndim: ", b.ndim)     # 2

c = np.array([[10],[20],[30]])  
print("c.shape:", c.shape)
````
