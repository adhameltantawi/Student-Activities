4- Add the following columns:

a. `Passed`

Rules:
- Grade >= 60 → `"Yes"`
- Otherwise → `"No"`

b. `Grade_Level`

Rules:
- Grade >= 90 → `"Excellent"`
- Grade >= 80 → `"Good"`
- Otherwise → `"Average"`

Then change the department of `"Omar"` from `"CS"` to `"AI"` without recreating the DataFrame.

````python
# a. Add 'Passed' column
df['Passed'] = df['Grade'].apply(lambda x: "Yes" if x >= 60 else "No")

# b. Add 'Grade_Level' column
def get_grade_level(grade):
    if grade >= 90:
        return "Excellent"
    elif grade >= 80:
        return "Good"
    else:
        return "Average"

df['Grade_Level'] = df['Grade'].apply(get_grade_level)

# Change the department of 'Omar' from 'CS' to 'AI'
df.loc[df['Name'] == 'Omar', 'Department'] = 'AI'

display(df)
````

7- Create the following DataFrame and store it in `df_missing`


```
data = {
    "Name": ["Ali", "Sara", None, "Lina"],
    "Age": [20, None, 19, 22],
    "Grade": [85, 92, 78, None]
}
```

Then:

- Display missing values in each column  
- Display total missing values  
- Fill missing names with `"Unknown"`  
- Fill numeric missing values with the column average  

````python
data = {
    "Name": ["Ali", "Sara", None, "Lina"],
    "Age": [20, None, 19, 22],
    "Grade": [85, 92, 78, None]
}
df_missing = pd.DataFrame(data)

print("Missing values in each column:")
display(df_missing.isna().sum())

print(f"\nTotal missing values: {df_missing.isna().sum().sum()}")

# Fill missing names with 'Unknown'
df_missing['Name'] = df_missing['Name'].fillna("Unknown")

# Fill numeric missing values with the column average
for col in ['Age', 'Grade']:
    df_missing[col] = df_missing[col].fillna(df_missing[col].mean())

print("\nDataFrame after filling missing values:")
display(df_missing)
````
