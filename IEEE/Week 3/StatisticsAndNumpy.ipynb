# Question 5:
Given the following dataset of exam scores:

65, 70, 72, 78, 80, 85, 90, 92, 95

What is the Interquartile Range (IQR)?

**A.** 15  
**B.** 20  
**C.** 25



````python
#The answer:
data is sorted and n = 6
Q1 is the median of the lower half (65,70,72,78) = (70+72)/2 = 71
Q3 is the median of the upper half (85,90,92,95) = (90+92)/2 = 91
IQR = Q3 - Q1 = 91 - 71 = 20
````

````python
The empirical rule (68–95–99.7 rule) states that for a normal distribution:

~68% of data falls within ±1 SD of the mean
~95% within ±2 SD
~99.7% within ±3
````

Write a function `check_empirical_rule(data)` that takes a list of numbers.

### Requirements

1. Calculate the **mean** from scratch (do not use built-in statistics libraries).
2. Calculate the **population standard deviation** from scratch.
3. Count exactly how many data points fall within:
   - ±1 standard deviation
   - ±2 standard deviations
   - ±3 standard deviations
4. Convert those counts into **percentages of the total dataset**.
5. Return a **dictionary** containing the calculated percentages so you can evaluate how close your dataset is to a "perfect" normal distribution (Empirical Rule: 68%–95%–99.7%).

### Expected Output Format (Example)

```python
{
    "within_1_std": 68.2,
    "within_2_std": 95.1,
    "within_3_std": 99.5
}

````python
def check_empirical_rule(data):
    n = len(data)
    mean = sum(data) / n
    std = (sum((x - mean) ** 2 for x in data) / n) ** 0.5
    def pct(k):
        count = sum(1 for x in data if abs(x - mean) <= k * std)
        return round(count / n * 100, 2)
    return {
        "within_1_std": pct(1),
        "within_2_std": pct(2),
        "within_3_std": pct(3),
    }
````

# The Outlier Hunter

Write a function `find_outliers(data)` that takes an **unsorted list of numerical values** and returns a list of all outliers.

## Requirements

Your function must:

1. Sort the data.
2. Mathematically compute:
   - The **Median**
   - The **First Quartile ($Q1$)**
   - The **Third Quartile ($Q3$)**
   - The **Interquartile Range (IQR)**
3. Calculate the lower and upper fences using the formula:

$$\text{Lower Fence} = Q1 - (1.5 \times \text{IQR})$$

$$\text{Upper Fence} = Q3 + (1.5 \times \text{IQR})$$

4. Return all numbers from the **original dataset** that fall strictly outside these fences.

## Expected Behavior

The function should return a list containing only the detected outliers.

````python
def find_outliers(data):
    s = sorted(data)
    n = len(s)

    def median(lst):
        m = len(lst) // 2
        return (lst[m - 1] + lst[m]) / 2 if len(lst) % 2 == 0 else lst[m]

    q1 = median(s[:n // 2])
    q3 = median(s[(n + 1) // 2:])
    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    return [x for x in data if x < lower or x > upper]
````

- Calculate the mean, median, and standard deviation of a given dataset(exam_scores) using NumPy.
- Find the range of a dataset (difference between the maximum and minimum values).
- Determine the mode of a dataset using NumPy.

````python
import numpy as np

np.random.seed(42)
exam_scores = np.random.randint(0, 101, 50)
print("Exam Scores:\n", exam_scores)

mean   = np.mean(exam_scores)
median = np.median(exam_scores)
std    = np.std(exam_scores)
range_ = np.max(exam_scores) - np.min(exam_scores)

values, counts = np.unique(exam_scores, return_counts=True)
mode = values[counts == counts.max()]

print(f"\nMean:   {mean}")
print(f"Median: {median}")
print(f"Std:    {std:.2f}")
print(f"Range:  {range_}")
print(f"Mode:   {mode}")
````
