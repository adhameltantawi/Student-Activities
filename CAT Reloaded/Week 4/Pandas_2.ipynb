### Group the dateframe by `Country` for the following questions:



````python
country_group = df.groupby('Country')
````

### 1. Show value counts for `Student` in the `United Kingdom`

#### a. How many are full-time students ?
#### b. How many are not students ?

````python
# Value counts for Student in United Kingdom
uk_student_counts = country_group['Student'].value_counts()['United Kingdom']
uk_student_counts
````

````python
# a. How many are full-time students?
full_time_students = uk_student_counts['Yes, full-time']
print(f'Full-time students in UK: {full_time_students}')
````

````python
# b. How many are not students?
not_students = uk_student_counts['No']
print(f'Not students in UK: {not_students}')
````

### 2. What's the median sallary `ConvertedComp` for developers in `Germany`,  `Italy`, `Finland`

````python
# Median salary for Germany, Italy, Finland
median_salary = country_group['ConvertedComp'].median()
median_salary[['Germany', 'Italy', 'Finland']]
````

````python
print(f"Germany  median salary : {median_salary['Germany']}")
print(f"Italy    median salary : {median_salary['Italy']}")
print(f"Finland  median salary : {median_salary['Finland']}")
````

### 3. Show the number of respondents from each country

````python
# Number of respondents from each country
respondents_per_country = country_group['Respondent'].count()
respondents_per_country
````

### 4. Show the number of respondents that worked with `Android` from each country mentioned in column `PlatformWorkedWith`

````python
# Number of respondents that worked with Android from each country
android_per_country = country_group['PlatformWorkedWith'].apply(
    lambda x: x.str.contains('Android', na=False).sum()
)
android_per_country
````

### 5. the function **`pandas.concat()`** to add these two series together in a new DataFrame

````python
# Concatenate respondents_per_country and android_per_country into one DataFrame
combined_df = pd.concat([respondents_per_country, android_per_country], axis=1)
combined_df.columns = ['Total_Respondents', 'Android_Users']
combined_df
````

### 6. Add a new column contains the percentage of people that worked with `Android` from each country

````python
# Add percentage column
combined_df['Android_Percentage'] = (combined_df['Android_Users'] / combined_df['Total_Respondents']) * 100
combined_df
````

### 1. remove nan values

````python
# write code
df.dropna(subset=['Age'], inplace=True)
````

### 2. make sure there is not nan values

````python
# write code
df['Age'].isna().sum()
````

### 3. clean `Age` Column
> **`hint:`** use .unique() and .replace()

````python
# write code
# First, inspect the unique values in the Age column
print(df['Age'].unique())

# Replace the verbose age range strings with cleaner labels
df['Age'] = df['Age'].replace({
    'Under 18 years old'  : '<18',
    '18 - 24 years old'   : '18-24',
    '25 - 34 years old'   : '25-34',
    '35 - 44 years old'   : '35-44',
    '45 - 54 years old'   : '45-54',
    '55 - 64 years old'   : '55-64',
    '65 years or older'   : '65+',
    'Prefer not to say'   : 'Unknown'
})

print(df['Age'].unique())
````
